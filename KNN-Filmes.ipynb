```python
# Importar as bibliotecas
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
```

```python
# Carregar conjunto de dados
df = pd.read_csv("filmes.csv", index_col=0)
df.head()
```

```python
# Remover Coluna Label
df = df.drop("Label", axis=1)
df = df.drop("Movie Name", axis=1)
df.head()
```

```python
df
```

```python
# Preparar o algoritmo
knn = NearestNeighbors(n_neighbors=3)
knn.fit(df)
```

```python
# Gostei do Forest Gump (97)
# Recomenda-se o filme?
knn.kneighbors([[8.8,0,1,0,0,0,0,0]])
```

```python
# Gostei do Spirited Away (65)
# Recomenda-se o filme?
knn.kneighbors([[8.6,0,0,0,0,0,0,0]])
```
